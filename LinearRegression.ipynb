```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
```

# Make data

```python
data = {
    'hours' : [2.5,3.5,9,10,1,3,2.5,6,7.5,6,7.5,
              1.5,2.5,1,7.5,4,12,5,5.5,5.5,7,10,
              3.5,6,7,4.5,12,10,2,9.5],
    'scores': [24,37,92,95,15,32,23,65,78,65,79,
             13,22,9,77,39,96,54,58,55,70,99,
             33,59,69,45,95,93,20,92]
}
dataframe = pd.DataFrame(data)
```

```python
X = dataframe['hours'].values.reshape(-1, 1)
y = dataframe['scores'].values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

# Train linear regression

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train,y_train)
```

```python
pred = linear_regressor.predict(X_test)
predictions = pd.DataFrame({'Actual' : y_test.squeeze(), 'Predicted': pred.squeeze()})
predictions
```

# Get MAE

```python
mae = mean_absolute_error(y_test,pred)
print(f'Mean Absolute Error: ${mae:.2f}')
```

```python
X_test.squeeze()
```

# Plot regression

```python
dataframe.plot.scatter(x='hours',y='scores',title='Hours and Scores')
coef = linear_regressor.coef_
inter = linear_regressor.intercept_
plt.plot(X,X*coef+inter)
```
