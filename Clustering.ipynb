```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
from sklearn.cluster import KMeans
from sklearn import datasets 
```

# Load data

```python
iris = datasets.load_iris()
X = iris.data 
y = iris.target
```

Apply K-Means

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X) 
```

# Plot in 3d Space

```python
fig = plt.figure()
subfigure = fig.add_subplot(111,projection="3d",elev=48,azim=134)
labels = kmeans.labels_
subfigure.scatter(X[:,3],X[:,0],X[:,2],c=labels.astype(float), edgecolor='k')
```
